```python
import cv2
import matplotlib.pyplot as plt
```

# Load Image

```python
sample_image = cv2.imread("data/daylight.jpg")

plt.imshow(sample_image)
plt.show()

# cv2.imshow("sample image", sample_image)
# cv2.waitKey(0)
```

# Resize Image

```python
sample_image.shape
```

```python
# resize with ratio
RESIZE_RATIO = 0.5

resized_sample_image = cv2.resize(sample_image, (0, 0), fx=RESIZE_RATIO, fy=RESIZE_RATIO)

plt.imshow(resized_sample_image)

print(resized_sample_image.shape)
```

```python
# resize directly

resized_sample_image2 = cv2.resize(sample_image, (480, 360))

plt.imshow(resized_sample_image2)

print(resized_sample_image2.shape)
```

# Convert Color

```python
# to grayscale

sample_image_gray = cv2.cvtColor(sample_image, cv2.COLOR_BGR2GRAY)


plt.imshow(sample_image_gray)
```

```python
# cv2.imshow("sample image gray", sample_image_gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

# Drawing on image

## Text

```python
height, width = resized_sample_image.shape[:2]
sample_image_with_text = cv2.putText(resized_sample_image.copy(), "Hello OpenCV!", (width // 2, height // 2), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 2)

plt.imshow(sample_image_with_text)
```

## Bounding box

```python
sample_image_with_bbox = cv2.rectangle(resized_sample_image.copy(), (100, 100), (400, 400), (0, 255, 0), 2)

plt.imshow(sample_image_with_bbox)
```

## Line

```python
sample_image_with_line = cv2.line(resized_sample_image.copy(), (0, 300), (800, 300), (0, 255, 0), 3)

plt.imshow(sample_image_with_line)
```
